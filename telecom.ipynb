# Analyzing the State Variable

```python
data.head(2)
```

```python
data.state.unique()
```

```python
data["state"].value_counts()
```

```python
plt.figure(figsize=(20,9))
sns.countplot(data["state"]);
plt.xticks(fontsize=13)
plt.ylabel("states",size=20)
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(20,7))
sns.countplot(x=data["state"], hue=data["churn"])
plt.show()
```

```python
a= data.groupby("state")["churn"].mean()*100
a.sort_values()
```

```python
X = data["state"].unique()
Y = data.groupby("state")["churn"].mean()

sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (20, 7)
plt.rcParams["lines.linestyle"] = "-"
plt.plot(X, Y, color="black", marker="*", markersize=10)

plt.xlabel("State")
plt.ylabel("Churn Rate")
plt.title("State-wise Churn Rate")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
a= data.groupby("state")["churn"].mean().sort_values(ascending=False).head(5)*100
print(a)
```

```python
c = pd.crosstab(data['state'], data['churn'], margins=True)
c['percentage_churns'] = (c[True] / c['All']) * 100

print(c.head())
print(type(c))
```

```python
c.info()
```

```python
sorted_c = c.sort_values(by='percentage_churns', ascending=False)
top_rows = sorted_c.head()

print(top_rows)
```

## this show the top 5 states with churn rate
